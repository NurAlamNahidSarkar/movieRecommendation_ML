```python
credits = pd.read_csv("/kaggle/input/the-movies-dataset/credits.csv")
keywords = pd.read_csv("/kaggle/input/the-movies-dataset/keywords.csv")
links = pd.read_csv("/kaggle/input/the-movies-dataset/links.csv")
links_small = pd.read_csv("/kaggle/input/the-movies-dataset/links_small.csv")
movies_metadata = pd.read_csv("/kaggle/input/the-movies-dataset/movies_metadata.csv")
ratings = pd.read_csv("/kaggle/input/the-movies-dataset/ratings.csv")
ratings_small = pd.read_csv("/kaggle/input/the-movies-dataset/ratings_small.csv")
```

```python
print(credits.columns)
print(keywords.columns)
print(links.columns)
print(links_small.columns)
print(movies_metadata.columns)
print(ratings.columns)
print(ratings_small.columns)
```

```python
credits.info()
keywords.info()
links.info()
links_small.info()
movies_metadata.info()
ratings.info()
ratings_small.info()
```

```python
print("--- Checking for 100% Identical Rows ---")

print(f"credits: {credits.duplicated().sum()}")
print(f"keywords: {keywords.duplicated().sum()}")
print(f"ratings: {ratings.duplicated().sum()}")
print(f"ratings_small: {ratings_small.duplicated().sum()}")
print(f"links: {links.duplicated().sum()}")
print(f"links_small: {links_small.duplicated().sum()}")
print(f"movies_metadata: {movies_metadata.duplicated().sum()}")
```

```python
print("\n--- Checking for Duplicate Keys ---")

print(f"movies_metadata 'id' duplicates: {movies_metadata.duplicated(subset=['id']).sum()}")

print(f"links 'movieId' duplicates: {links.duplicated(subset=['movieId']).sum()}")
print(f"links 'tmdbId' duplicates: {links.duplicated(subset=['tmdbId']).sum()}")
print(f"links_small 'movieId' duplicates: {links_small.duplicated(subset=['movieId']).sum()}")
print(f"links_small 'tmdbId' duplicates: {links_small.duplicated(subset=['tmdbId']).sum()}")

print(f"ratings duplicates: {ratings.duplicated(subset=['userId', 'movieId']).sum()}")
print(f"ratings_small duplicates: {ratings_small.duplicated(subset=['userId', 'movieId']).sum()}")

print(f"credits duplicates: {credits.duplicated(subset=['id']).sum()}")
print(f"keywords duplicates: {keywords.duplicated(subset=['id']).sum()}")
```

```python
credits = credits.drop_duplicates()
print(f"credits: {credits.duplicated().sum()}")
credits = credits.drop_duplicates(subset=['id'])
print(f"credits duplicates: {credits.duplicated(subset=['id']).sum()}")
```

```python
keywords = keywords.drop_duplicates()
print(f"keywords: {keywords.duplicated().sum()}")
print(f"keywords duplicates: {keywords.duplicated(subset=['id']).sum()}")
```

```python
movies_metadata = movies_metadata.drop_duplicates(subset=['id'])
print(f"movies_metadata: {movies_metadata.duplicated().sum()}")
print(f"movies_metadata duplicates: {movies_metadata.duplicated(subset=['id']).sum()}")
```

```python
links = links.drop_duplicates(subset=['tmdbId'])
print(f"links: {links.duplicated().sum()}")
print(f"links duplicates: {links.duplicated(subset=['tmdbId']).sum()}")
```

```python
links_small  = links_small .drop_duplicates(subset=['tmdbId'])
print(f"links_small : {links_small .duplicated().sum()}")
print(f"links_small  duplicates: {links_small .duplicated(subset=['tmdbId']).sum()}")
```

```python
all_ratings = pd.concat([ratings, ratings_small])
```

```python
all_ratings.shape
```

```python
all_links = pd.concat([links,links_small])
```

```python
all_links.shape
```

```python
credits_keywords = pd.concat([credits,keywords])
```

```python
credits_keywords.shape
```

```python
movies_metadata['id'] = pd.to_numeric(movies_metadata['id'], errors='coerce')

# 2. Now, convert from float64 to Int64 (which allows <NA>)
movies_metadata['id'] = movies_metadata['id'].astype('Int64')
```

```python
movies_metadata.info()
```

```python
movies_metadata.shape
```

```python
movies = movies_metadata.merge(credits_keywords,on='id')
```

```python
movies.shape
```

```python
movies.head()
```

```python
movies = movies.drop_duplicates()
print(f"movies: {movies.duplicated().sum()}")
movies = movies.drop_duplicates(subset=['id'])
print(f"movies duplicates: {movies.duplicated(subset=['id']).sum()}")
```

```python
movies = movies.merge(all_links,left_on='id',right_on='tmdbId')
```
